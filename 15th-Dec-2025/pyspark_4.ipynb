```python

orders_csv = (spark.read
                  .option("header", True)
                  .option("inferSchema", True)
                  .csv("orders_csv/"))

filtered_csv = orders_csv.filter(F.col("order_amount") > 700)
filtered_csv.show(truncate=False)
```

```python

sorted_csv = orders_csv.orderBy(F.col("delivery_time_minutes").desc())
(sorted_csv.write
    .mode("overwrite")
    .option("header", True)
    .csv("orders_csv_sorted/"))
```

```python

delivered = df.filter(F.col("order_status") == "Delivered")
(delivered.write
    .mode("overwrite")
    .json("delivered_orders_json/"))
```

```python

delivered_json = spark.read.json("delivered_orders_json/")
mumbai_card = delivered_json.filter(
    (F.col("city") == "Mumbai") & (F.col("payment_mode") == "Card")
)
mumbai_card.show(truncate=False)
```

```python

with_category = delivered_json.withColumn(
    "delivery_category",
    F.when(F.col("delivery_time_minutes") > 45, F.lit("Late")).otherwise(F.lit("OnTime"))
)

(with_category.write
    .mode("overwrite")
    .json("delivered_orders_with_category_json/"))
```

```python

(with_category.coalesce(1)
    .write.mode("overwrite")
    .json("delivered_orders_with_category_singlepart_json/"))
```

```python

by_cuisine_revenue = (df.groupBy("cuisine")
                        .agg(F.sum("order_amount").alias("total_order_amount"))
                        .orderBy(F.col("total_order_amount").desc()))
by_cuisine_revenue.show(truncate=False)

# Top cuisine:
top_cuisine = by_cuisine_revenue.first()
print("Top cuisine:", top_cuisine["cuisine"], "with", top_cuisine["total_order_amount"])
```

```python

by_city_count = (df.groupBy("city")
                   .agg(F.count("*").alias("total_orders"))
                   .orderBy(F.col("total_orders").desc()))
by_city_count.show(truncate=False)

top_city = by_city_count.first()
print("Top city:", top_city["city"], "with", top_city["total_orders"], "orders")
```

```python

by_payment = (df.groupBy("payment_mode")
                .agg(F.count("*").alias("usage_count"))
                .orderBy(F.col("usage_count").desc()))
by_payment.show(truncate=False)

top_payment = by_payment.first()
print("Most used payment mode:", top_payment["payment_mode"], "with", top_payment["usage_count"])
```

```python

(df.repartition(4)
   .write.mode("overwrite")
   .parquet("orders_parquet_4parts/"))
```

```python

report = (df.groupBy("city")
            .agg(
                F.count("*").alias("total_orders"),
                F.sum("order_amount").alias("total_revenue")
            )
            .orderBy(F.col("total_revenue").desc()))

report.show(truncate=False)

(report.write
    .mode("overwrite")
    .parquet("city_report_parquet/"))
```
