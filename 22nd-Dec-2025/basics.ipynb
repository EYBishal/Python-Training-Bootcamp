```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkByExamples.com").getOrCreate()
```

```python
data=[
    ("James","Smith","USA","CA",19),
    ("Michael","Rose","USA","NY",1),
    ("Robert","Williams","USA","CA",3),
    ("Maria","Jones","USA","FL",4)
]
column=["firstname","lastname","country","state","amount"]
df=spark.createDataFrame(data=data,schema=column)
df.show(truncate=False)
```

```python
from pyspark.sql.functions import col
df.withColumn("tot",col("amount")*7).show()
```

```python
df.withColumnRenamed("firstname","realname").show()
```

```python
df.replace("CA","California").show()
```

```python
from pyspark.sql.functions import when
df.withColumn("country",when(df.country=="USA","United States").otherwise(df.country)).show()
```
