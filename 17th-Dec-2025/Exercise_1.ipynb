**Predicate pushdown** means letting the data source apply filters during read, reducing data transferred to Spark.

```python
df_c=df_orders.select("order_id","region", "city","restaurant_id","food_item","amount")
df_filter=df_c.filter(col("region")=="North")
df_final=df_filter.withColumn("amount_in_tens",col("amount")/10)
df_final.explain(True)
df_final.count()
df_final.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank,col

w = Window.partitionBy("region").orderBy(col("order_date"))

df_out = df_orders.withColumn("running_total", f.sum(col("amount")).over(w))
df_out.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank

w = Window.partitionBy("region").orderBy(col("amount").desc())

df_out = df_orders.withColumn("running_total", f.rank().over(w))
df_out.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank

w = Window.partitionBy("restaurant_id").orderBy(col("delivery_time_min").desc())

df_out = df_orders.withColumn("running_total", f.row_number().over(w))
df_out.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank

w = Window.partitionBy("food_item").orderBy(col("region").desc())

df_out = df_orders.withColumn("running", f.dense_rank().over(w))
df_out.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank

w = Window.partitionBy("region").orderBy(col("amount").desc())

df_out = df_orders.withColumn("running_total", f.rank().over(w)).filter(col("running_total")<=2)
df_out.show()
```

```python

from pyspark.sql import functions as F
from pyspark.sql.window import Window

w = Window.partitionBy("restaurant_id").orderBy("delivery_time_min")

df2 = df_orders.withColumn(
    "cumulative_time",
    F.sum("delivery_time_min").over(w)
)

df2.show()
```
