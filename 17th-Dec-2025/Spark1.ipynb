```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, rank, row_number, dense_rank
w=  Window.partitionBy("region").orderBy(col("amount").desc())
df_out=df_sales.withColumn("rank",f.dense_rank().over(w)).filter(col("rank")<=2)
df_out.show()
```

**rank()**
Assigns the same rank to tied values, but skips numbers after a tie.
Example: 1, 1, 3, 4 (notice it skipped 2 because two rows shared rank 1).


**dense_rank()**
Assigns the same rank to tied values, but does NOT skip numbers.
Example: 1, 1, 2, 3 (no gaps).

```python

from pyspark.sql.window import Window
from pyspark.sql.functions import col, sum as F_sum

w = Window.partitionBy("region").orderBy(col("sale_date").asc()) \
          .rowsBetween(Window.unboundedPreceding, Window.currentRow)

df_win = df_sales.withColumn("running_total", F_sum(col("amount")).over(w))
df_win.explain(True)
```

**Why window functions introduce sorting**


A window spec like Window.partitionBy("region").orderBy("sale_date") requires two things:

Partitioning by region: rows of the same region must be grouped. Spark typically performs an Exchange (shuffle) to repartition by the partition key.
Ordering by sale_date within each region: to compute running totals, ranks, first/last, etc., rows must be in a deterministic order. This introduces a Sort phase before the Window operator computes the frame (e.g., from the start of the partition to the current row).
